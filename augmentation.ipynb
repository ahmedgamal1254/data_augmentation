```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
imgs=os.listdir('/content/drive/MyDrive/cats')
print(imgs)
```

```python
# path_img=os.path.join('/content/drive/MyDrive/cats',imgs[1])
imgs_arr=[]
imgs2=['16.jpg', '17.jpg', '18.jpg', '19.jpg', '20.jpg']
for i in imgs2:
  path=os.path.join('/content/drive/MyDrive/cats',i)
  arr=plt.imread(path)
  arr=cv2.resize(arr,dsize=(200,200))
  print(np.shape(arr))
  imgs_arr.append(arr)
```
