```python
import os
import glob

import cv2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, 
    channel_shift_range=10., horizontal_flip=True)
```

```python
imgs=os.listdir('/content/drive/MyDrive/cats')
print(imgs)
```

```python
img_path=os.path.join('/content/drive/MyDrive/cats',imgs[0])
print(img_path)
```

```python
img_arr=plt.imread(img_path)
plt.imshow(img_arr)
```

```python
image = np.expand_dims(plt.imread(img_path),0)
```

```python
aug_iter = gen.flow(image,save_to_dir='/content/drive/MyDrive/aug',save_prefix='cat_',save_format='jpeg')
```

```python
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```
