```python
import os
import glob

import numpy as np
import matplotlib.pyplot as plt

import imgaug.augmenters as iaa
```

```python
import cv2
```

```python
imgs=os.listdir('/content/drive/MyDrive/cats')
imgs
```

```python
arr_imgs=[]
for i in imgs:
  img=os.path.join('/content/drive/MyDrive/cats',i)
  arr=plt.imread(img)
  new_arr=cv2.resize(arr,dsize=(200,200))
  arr_imgs.append(new_arr)
```

```python
len(arr_imgs)
```

```python
np.shape(arr_imgs)
```

```python
plt.imshow(arr_imgs[4])
```

```python
# make augmentation for data
aug=iaa.Sequential([
    iaa.Fliplr(1)
])

aug_img=aug(images=arr_imgs)
```

```python
for i in range(len(aug_img)):
  img=plt.imshow(aug_img[i])
  plt.savefig(f'/content/drive/MyDrive/cats/cat_{imgs[i]}.jpg')
  plt.show()
```
