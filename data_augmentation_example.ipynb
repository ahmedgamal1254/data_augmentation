```python
import os
import glob

import numpy as np
import matplotlib.pyplot as plt

import cv2

import imgaug.augmenters as iaa
```

```python
path=os.listdir('/content/drive/MyDrive/imgs')
path
```

```python
imgs_arr=[]
for i in path:
  img=os.path.join('/content/drive/MyDrive/imgs',i)
  arr=plt.imread(img)
  new_arr=cv2.resize(arr,dsize=(200,200))
  imgs_arr.append(new_arr)
```

```python
plt.imshow(imgs_arr[0])
```

```python
np.shape(imgs_arr)
```

```python
aug=iaa.Sequential([
    iaa.Fliplr(1)
])

aug_img=aug(images=imgs_arr)
```

```python
len(aug_img)
```

```python
os.chdir('/content/sample_data')
```

```python
aug2=iaa.Sequential([
    iaa.Crop(px=(0, 16)),
    iaa.GaussianBlur(sigma=(0.4, 7.0))
])

aug_img2=aug2(images=imgs_arr)
```

```python
for i in aug_img2:
  img=plt.imshow(i)
  plt.show()
```

```python
for i in aug_img:
  img=plt.imshow(i)
  plt.savefig(f'{np.random.randint(50)}.png')
  plt.show()
```
